Question 3 C#
Implement function triangleArea(a,b,c) that takes as input the lengths of the 3
sides of a triangle and returns the area of the triangle. By Heron's formula, the area
of a triangle with side lengths a, b, and c is s(s − a)(s − b)(s − c), where
s = (a + b + c)/2.
>>> triangleArea(2,2,2)
1.7320508075688772

```python
# Answer 3

import math

def triangleArea(a, b, c):
    # Calculate the semi-perimeter
    s = (a + b + c) / 2
    # Calculate the area using Heron's formula
    area = math.sqrt(s * (s - a) * (s - b) * (s - c))
    return area

# Example :
    
area = triangleArea(2, 2, 2)
print(area)

# Output: 1.7320508075688772
```

Write a program in C# Sharp to separate odd and even integers in separate arrays.
Go to the editor
Test Data :
Input the number of elements to be stored in the array :5
Input 5 elements in the array :
element - 0 : 25
element - 1 : 47
element - 2 : 42
element - 3 : 56
element - 4 : 32
Expected Output:
The Even elements are:
42 56 32
The Odd elements are :
25 47

```python
using System;

class Program
{
    static void Main()
    {
        # Declare the array size and elements
        int[] arr = new int[5];
        int[] evenNumbers = new int[5];  // Array for even numbers
        int[] oddNumbers = new int[5];   // Array for odd numbers
        int evenCount = 0, oddCount = 0; // Counters for even and odd numbers

        # Taking input from the user
        Console.WriteLine("Input 5 elements in the array:");
        for (int i = 0; i < arr.Length; i++)
        {
            Console.Write("element - {0} : ", i);
            arr[i] = Convert.ToInt32(Console.ReadLine());

            # Separate even and odd numbers
            if (arr[i] % 2 == 0)
            {
                evenNumbers[evenCount] = arr[i];
                evenCount++;
            }
            else
            {
                oddNumbers[oddCount] = arr[i];
                oddCount++;
            }
        }

        # Print the even numbers
        Console.WriteLine("The Even elements are:");
        for (int i = 0; i < evenCount; i++)
        {
            Console.Write(evenNumbers[i] + " ");
        }

        # Print the odd numbers
        Console.WriteLine("\nThe Odd elements are:");
        for (int i = 0; i < oddCount; i++)
        {
            Console.Write(oddNumbers[i] + " ");
        }
    }
}
```

```python
#Answer 5 
#Part a: Function inside(x, y, x1, y1, x2, y2)
#This function will check if the point (x, y) lies inside the rectangle defined by the lower-left corner (x1, y1) and the upper-right corner (x2, y2).
using System;

class Program
{
    # Function to check if a point is inside the rectangle
    static bool Inside(double x, double y, double x1, double y1, double x2, double y2)
    {
        # Return true if the point (x, y) lies within the rectangle boundaries
        return x >= x1 && x <= x2 && y >= y1 && y <= y2;
    }

    static void Main()
    {
        # Test cases for part a
        Console.WriteLine(Inside(1, 1, 0, 0, 2, 3));  # Output: True
        Console.WriteLine(Inside(-1, -1, 0, 0, 2, 3)); # Output: False
    }
}
```

```python
# Part b: point (1, 1) lies in both rectangles
using System;

class Program
{
    # Function to check if a point is inside the rectangle, returns 1 for True and 0 for False
    static int Inside(double x, double y, double x1, double y1, double x2, double y2)
    {
        return (x1 <= x && x <= x2) && (y1 <= y && y <= y2) ? 1 : 0;
    }

    static void Main()
    {
        # Define the point to check
        double pointX = 1;
        double pointY = 1;

        # Check if the point is inside both rectangles
        int inRectangle1 = Inside(pointX, pointY, 0.3, 0.5, 1.1, 0.7);
        int inRectangle2 = Inside(pointX, pointY, 0.5, 0.2, 1.1, 2);

        # Check if the point is inside both rectangles
        if (inRectangle1 + inRectangle2 == 2)
        {
            Console.WriteLine("True"); # The point is inside both rectangles
        }
        else
        {
            Console.WriteLine("False"); # The point is not inside both rectangles
        }
    }
}
```

16. You can turn a word into pig-Latin using the following two rules (simplified):
• If the word starts with a consonant, move that letter to the end and append
'ay'. For example, 'happy' becomes 'appyhay' and 'pencil' becomes 'encilpay'.
• If the word starts with a vowel, simply append 'way' to the end of the word.
For example, 'enter' becomes 'enterway' and 'other' becomes 'otherway' . For
our purposes, there are 5 vowels: a, e, i, o, u (so we count y as a consonant).
Write a function pig() that takes a word (i.e., a string) as input and returns its pig-
Latin form. Your function should still work if the input word contains upper case
characters. Your output should always be lower case however.
>>> pig('happy')
'appyhay'
>>> pig('Enter')
'enterway'

```python
#Answer 6

def pig(word):
    # Convert the word to lowercase
    word = word.lower()
    
    # Define vowels
    vowels = 'aeiou'
    
    # Check if the first letter is a vowel
    if word[0] in vowels:
        return word + 'way'
    else:
        # If the first letter is a consonant, move it to the end and add 'ay'
        return word[1:] + word[0] + 'ay'

# Test cases
print(pig('happy'))  # Output: 'appyhay'
print(pig('Enter'))  # Output: 'enterway'


#Example outputs:

>>> pig('happy')
'appyhay'
>>> pig('Enter')
'enterway'
```

Question 7 Python
File bloodtype1.txt records blood-types of patients (A, B, AB, O or OO) at a clinic.
Write a function bldcount() that reads the file with name name and reports (i.e.,
prints) how many patients there are in each bloodtype.
>>> bldcount('bloodtype.txt')
There are 10 patients of blood type A.
There is one patient of blood type B.
There are 10 patients of blood type AB.
There are 12 patients of blood type O.
There are no patients of blood type OO.

```python
#Answer 7 

def bldcount(filename):
    # Dictionary to store blood type counts
    blood_types = {'A': 0, 'B': 0, 'AB': 0, 'O': 0, 'OO': 0}
    
    # Open and read the file
    with open(filename, 'r') as file:
        data = file.read().split()  # Split the file content into a list of blood types
    
    # Count the occurrences of each blood type
    for blood_type in data:
        if blood_type in blood_types:
            blood_types[blood_type] += 1
    
    # Print the result
    for blood_type, count in blood_types.items():
        if count == 0:
            print(f"There are no patients of blood type {blood_type}.")
        elif count == 1:
            print(f"There is one patient of blood type {blood_type}.")
        else:
            print(f"There are {count} patients of blood type {blood_type}.")

# Example:
# bldcount('bloodtype.txt')
```

Encryption is the process of hiding the meaning of a text by substituting letters in the
message with other letters, according to some system. If the process is successful, no
one but the intended recipient can understand the encrypted message. Cryptanalysis
refers to attempts to undo the encryption, even if some details of the encryption are
unknown (for example, if an encrypted message has been intercepted). The first step
of cryptanalysis is often to build up a table of letter frequencies in the encrypted text.
Assume that the string letters is already defined as
'abcdefghijklmnopqrstuvwxyz'. Write a function called frequencies()
that takes a string as its only parameter, and returns a list of integers, showing the
number of times each character appears in the text. Your function may ignore any
characters that are not in letters.
>>> frequencies('The quick red fox got bored and went home.')
[1, 1, 1, 3, 5, 1, 1, 2, 1, 0, 1, 0, 1, 2, 4, 0, 1, 2, 0, 2,
1, 0, 1, 1, 0, 0]
>>> frequencies('apple')

```python
#Answer 10

def frequencies(text):
    letters = 'abcdefghijklmnopqrstuvwxyz'
    # Initialize a list with 26 zeros (one for each letter)
    freq_list = [0] * 26
    
    # Convert the input text to lowercase
    text = text.lower()
    
    # Iterate through each character in the input text
    for char in text:
        if char in letters:
            # Find the index of the character in letters and increment the count
            index = letters.index(char)
            freq_list[index] += 1
    
    return freq_list

# Example 
print(frequencies('The quick red fox got bored and went home.'))
print(frequencies('apple'))
```
